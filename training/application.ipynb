```python
import tensorflow as tf
import numpy as np
newModel=tf.keras.models.load_model("../saved_models/1")
```

```python
import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from tkinter import filedialog as fd 
from tkinter import PhotoImage,Label
from PIL import ImageTk, Image
class_names=['Tomato Early Blight',
 'Tomato Late Blight',
 'Tomato Leaf Mold',
 'Tomato Septoria Leaf Spot',
 'Tomato Spider  Mites Two Spotted Spider Mite',
 'Tomato Target Spot',
 'Tomato Tomato YellowLeaf Curl Virus',
 'Tomato Tomato Mosaic Virus',
 'Tomato Healthy']
form=tk.Tk()
form.title("Tomato-Disease Classifier")
form.geometry("500x400+500+350")
form.minsize(450,400)
form.maxsize(550,550)
form.state("iconic")
label=tk.Label(form,text="")
label.pack()
predict_label=tk.Label(form,text="")
predict_label.pack()
confidence_label=tk.Label(form,text="")
confidence_label.pack()
def openfn():
    filename = fd.askopenfilename(title='open')
    return filename
def open_img():
    
    global label
    global predict_label
    global confidence_label
    x = openfn()
    label.destroy()
    predict_label.destroy()
    confidence_label.destroy()
    img = Image.open(x)
    img = img.resize((256, 256), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    label = Label(form, image=img)
    label.image = img
    label.pack()
    data=tf.keras.preprocessing.image.load_img(x, grayscale=False, color_mode="rgb", target_size=None, interpolation="nearest")
    img_batch=np.expand_dims(data,0)
    predictions=newModel.predict(img_batch)
    predicted_class=class_names[np.argmax(predictions[0])]
    confidence=round(100*(np.max(predictions)),2)
    predict_label=Label(form, text = predicted_class)
    predict_label.config(font=(_,20))
    predict_label.pack()
    confidence_label=Label(form,text="%"+str(confidence))
    confidence_label.config(font=(_,20))
    confidence_label.pack()
def select():
    
    open_img()
    
def quit():
    form.destroy()
button1=tk.Button(form,text="Select Image",fg="black",bg="red",command=select)
button2=tk.Button(form,text="Quit",fg="black",bg="red",command=quit)
button1.place(x=190, y=350)
button2.place(x=270, y=350)

form.mainloop()
```
